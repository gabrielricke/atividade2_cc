```python
# questao 9
peso = float(input("Digite seu peso em kg: "))
altura = float(input("Digite sua altura em metros: "))
imc = peso / (altura ** 2)
print(f"Seu IMC é: {imc:.2f}")
if imc < 18.5:
    print("Condição: Abaixo do peso")
elif 18.5 <= imc <= 24.9:
    print("Condição: Peso ideal (parabéns)")
elif 25.0 <= imc <= 29.9:
    print("Condição: Levemente acima do peso")
elif 30.0 <= imc <= 34.9:
    print("Condição: Obesidade grau I")
elif 35.0 <= imc <= 39.9:
    print("Condição: Obesidade grau II (severa)")
else:
    print("Condição: Obesidade grau III (mórbida)")
```

```python
# questao 12
valor = float(input("Digite o valor do produto: R$ "))
print("\nFormas de Pagamento:")
print("1 - À Vista em Dinheiro ou Pix (15% de desconto)")
print("2 - À Vista no Cartão de Crédito (10% de desconto)")
print("3 - Parcelado em 2x no Cartão (sem juros)")
print("4 - Parcelado em 3x ou mais no Cartão (10% de juros)")
codigo = int(input("Digite o código da forma de pagamento: "))
if codigo == 1:
    valor_final = valor * 0.85
    print(f"Valor com 15% de desconto: R$ {valor_final:.2f}")
elif codigo == 2:
    valor_final = valor * 0.90
    print(f"Valor com 10% de desconto: R$ {valor_final:.2f}")
elif codigo == 3:
    valor_final = valor
    parcela = valor_final / 2
    print(f"Pagamento em 2x de R$ {parcela:.2f} (sem juros)")
    print(f"Valor total: R$ {valor_final:.2f}")
elif codigo == 4:
    valor_final = valor * 1.10
    print(f"Valor com 10% de juros: R$ {valor_final:.2f}")
else:
    print("Código de pagamento inválido.")
```

```python
# questao 15
from datetime import datetime
ano_nascimento = int(input("Digite o ano de nascimento: "))
ano_atual = datetime.now().year
idade_anos = ano_atual - ano_nascimento
dias_totais = idade_anos * 365
anos = dias_totais // 365
resto_dias = dias_totais % 365
meses = resto_dias // 30
dias = resto_dias % 30
print(f"\nVocê viveu aproximadamente {anos} anos, {meses} meses e {dias} dias.")
```

```python
# questao 16
lado1 = float(input("Digite o comprimento do primeiro lado: "))
lado2 = float(input("Digite o comprimento do segundo lado: "))
lado3 = float(input("Digite o comprimento do terceiro lado: "))
if lado1 <= 0 or lado2 <= 0 or lado3 <= 0:
    print("Os lados do triângulo devem ser positivos.")
elif lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    if lado1 == lado2 == lado3:
        print("O triângulo é EQUILÁTERO.")
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print("O triângulo é ISÓSCELES.")
    else:
        print("O triângulo é ESCALENO.")
else:
    print("Os valores fornecidos não formam um triângulo válido.")
```

```python
# questao 18
altura_francisco = 1.50
altura_sara = 1.10
crescimento_francisco = 0.02  # 2 cm = 0.02 m
crescimento_sara = 0.03       # 3 cm = 0.03 m
anos = 0
while altura_sara <= altura_francisco:
    altura_francisco += crescimento_francisco
    altura_sara += crescimento_sara
    anos += 1
print(f"Sara será mais alta que Francisco em {anos} anos.")
```
